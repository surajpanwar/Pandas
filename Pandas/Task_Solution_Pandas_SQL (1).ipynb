# 10. Try to find out a third highest most selling dress id 

```python

import mysql.connector as connection
mydb = connection.connect(host="localhost",user="suraj", passwd="ThisisMyPassword",database="shop",use_pure=True,auth_plugin='mysql_native_password')

query = '''
 SELECT Dress_ID, SUM(`29/8/2013` + 
`31/8/2013` + 
`2/9/2013` + 
`4/9/2013` + 
`6/9/2013` + 
`8/9/2013` + 
`10/9/2013` + 
`12/9/2013` + 
`14/9/2013` + 
`16/9/2013` + 
`18/9/2013` + 
`20/9/2013` + 
`22/9/2013` + 
`24/9/2013` + 
`26/9/2013` + 
`28/9/2013` + 
`30/9/2013` + 
`2/10/2013` + 
`4/10/2013` + 
`6/10/2013` + 
`8/10/2013` + 
`10/10/2013` + 
`12/10/2013`) AS Total_sales from Dress_Sales
group by Dress_ID
order by Total_sales desc limit 1 offset 2;
'''

```
